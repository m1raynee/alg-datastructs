```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
# Plot experiment 4 results
with open("../../resources/experiment4.csv", "r") as f:
    lines = f.readlines()
    sizes = []
    avg_heights = []
    for line in lines[1:]:
        size, avg_height = line.strip().split(",")
        sizes.append(int(size))
        avg_heights.append(float(avg_height))

plt.plot(sizes, avg_heights, marker='o')

plt.title("BST: Средняя высота дерева от количества ключей")
plt.xlabel("Количество ключей")
plt.ylabel("Средняя высота дерева")
plt.grid(True)

# plt.xscale('log')
# plt.yscale('log')
```

```python
def make_plot(filename):
    with open(filename, "r") as f:
        lines = f.readlines()
        sizes = []
        heights = []
        for line in lines[1:]:
            size, height = line.strip().split(",")
            sizes.append(int(size))
            heights.append(int(height))
    
        log2_n = [np.log2(n) for n in sizes]
        a, b = np.polyfit(log2_n, heights, 1)
        print(f"Линейная регрессия для высоты дерева: height = {a:.4f} * log2(n) + {b:.4f}")

        plt.plot(sizes, heights, color='red', label='Высота дерева')
        plt.plot(sizes, log2_n, color='blue', linestyle='--')
        plt.plot(sizes, [a * x + b for x in log2_n], color='purple', label=f'Линейная регрессия ({a:.2f} log2(n) + {b:.2f})')

        plt.grid(True)
        plt.legend()

        plt.xlabel("Количество ключей (n)")
        plt.ylabel("Высота дерева (h)")

        return sizes, log2_n
```

```python
_, log2_n = make_plot("../../resources/experiment4_single.csv")

plt.title("BST: Высота дерева от количества ключей")
plt.show()
```

```python
x, log2_n = make_plot("../../resources/experiment5_avl.csv")
plt.plot(x, [1.44 * xi - 1.33 for xi in log2_n], linestyle='--', color='blue')
plt.title("AVL: Высота дерева от количества случайных ключей")
plt.show()
```

```python
x, log2_n = make_plot("../../resources/experiment5_rbt.csv")
plt.title("RBT: Высота дерева от количества случайных ключей")
plt.plot(x, [2 * np.log2(xi + 1) for xi in x], linestyle='--', color='blue')
plt.show()
```
